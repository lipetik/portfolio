```python
data = ascii.read("gep3-bulk.pdos_tot", comment='#')
data1 = ascii.read("Ge_d_total.dat", comment='#')
data2 = ascii.read("Ge_p_total.dat", comment='#')
data3 = ascii.read("Ge_s_total.dat", comment='#')
data4 = ascii.read("P_p_total.dat", comment='#')
data5 = ascii.read("P_p_total.dat", comment='#')
```

```python
plt.plot(data[0][:]+2.63, data[1][:], lw=2.3, color='red', label='Total')
plt.axvline(0, color='black')
plt.ylim(0,50)
plt.xlim(-2,2)
```

```python
plt.figure(figsize=(12,12))
plt.plot(data1[0][:]+2.6371, data1[1][:], lw=2.3, color='blue', label='Ge-s')
plt.plot(data2[0][:]+2.6371, data2[1][:], lw=2.3, color='magenta', label='Ge-p')
plt.plot(data3[0][:]+2.6371, data3[1][:], lw=2.3, color='yellow', label='Ge-d')
plt.plot(data4[0][:]+2.6371, data4[1][:], lw=2.3, color='green', label='P-s')
plt.plot(data5[0][:]+2.6371, data5[1][:], lw=2.3, color='turquoise', label='P-p')
#plt.xlim(-15,15)
plt.xlim(-3,3)
plt.ylim(0,50)
plt.legend()
plt.axvline(0, color='black')
```
